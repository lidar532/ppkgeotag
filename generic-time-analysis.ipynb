```python
#@title Plot Photo and Event Times { form-width: "25%", display-mode: "form" }
import pandas as pd
import numpy as np
import bokeh as bk
from bokeh.plotting import figure
from bokeh.io import output_notebook,  show
TSV_File_Name = "/content/jd197-for-plotting.txt" #@param {type:"string"}
output_notebook()
tools='pan,wheel_zoom,xpan, xwheel_zoom, box_zoom,reset,undo, redo, save,  crosshair, hover'
p = figure(width=1000, height=400, output_backend = "webgl", tools=tools,
            tooltips="@GPS_SOD{11111.111111}: @ImageName", title=TSV_File_Name,
           active_drag="xpan", active_scroll ='xwheel_zoom')
p.y_range = bk.models.Range1d(-50, 150)
p.xaxis.axis_label = 'GPS SOD'
data=pd.read_csv(TSV_File_Name, delimiter='\t')
p.scatter('GPS_SOD', 'CAM', source=data, marker='square', size=8,  color="blue", alpha=0.20)
show(p)
display(data[0:5])
```

```python
#@title Histogram of events vs Offset from each second { display-mode: "form" }
dt = np.fmod(data['GPS_SOD'], 1.0)
dt = np.where(  dt > 0.9, dt-1, dt)
y = np.histogram(dt, bins=1000)
htools='pan,wheel_zoom, box_zoom,reset,undo, redo, save'
h = figure(width=1000, height=400, 
                       output_backend = "webgl", tools=htools,title=TSV_File_Name, 
                      active_drag="pan", active_scroll ='wheel_zoom', 
                       tooltips="$x{1.111}: $y{1111}")
h.line(y[1][0:1000], y[0], width=.5)
h.circle(y[1][0:1000], y[0], size=5)
h.xaxis.axis_label='Offset from the second.'
h.yaxis.axis_label='Event count'
show(h)
```
